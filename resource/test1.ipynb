# pandas rolling and Expanding


```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('05-Pandas-with-Time-Series/time_data/walmart_stock.csv',index_col='Date',parse_dates=True)
```

```python
df.head()
```

```python
df['Open'].plot(figsize=(16,6))
```

```python
# averaging it out by mean

df.rolling(7).mean().head(14) df# we are caluting mean of every preivious 7 days 
```

```python
df['Open'].plot()
df.rolling(window=90).mean()['Close'].plot(figsize=(16,6))
```

```python
#defing legends
# make rolling value as column like done below

df['Close 30 Day Ma']=df['Close'].rolling(window=30).mean()# creating 30 days moving avergae columns
```

```python
df.head()
```

```python
# in plot selecting Close 30 Day Ma and Close column . plot will automatically give a legend name
df[['Close 30 Day Ma','Close']].plot(figsize=(16,6))
```

```python
# expanding method
df['Close'].expanding().mean().plot(figsize=(16,6))
```

```python
# close 30 MA
df['Close:20 day Mean']=df['Close'].rolling(20).mean()

# Upper= 20 MA+2* Std(20)
df['Upper']=df['Close:20 day Mean']+2*df["Close"].rolling(20).std()
# Lowe=20MA-2*Std(20)
df['Lower']=df['Close:20 day Mean']-2*df["Close"].rolling(20).std()
#close
df[['Close','Close:20 day Mean','Upper','Lower']].tail(200).plot(figsize=(16,6))
```
